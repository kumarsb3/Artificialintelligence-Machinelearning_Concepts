
# ** n8n **

Github link - https://github.com/n8n-io/n8n

n8n - It is an automation tool that lets you create custom workflows.
n8n uses a node-based architecture to define workflows. A workflow is a series of nodes connected in sequence to perform a task.

### Key Components
* Workflow: A sequence of steps to automate a task.
* Node: A single step in the workflow. It could be an API call, a webhook, or a simple logic operation.
* Trigger: A special type of node that starts the workflow (e.g., chat msg, webhook or time schedule).
* Execution: When a workflow runs.

### Installation
Install docker desktop and then install n8n via below docker commands

* docker volume create n8n_data  

* docker run -it --rm --name n8n -p 5678:5678 -v n8n_data:/home/node/.n8n docker.n8n.io/n8nio/n8n

* localhost:5678 - this is where the docker hosted n8n would run on local machine.


### Workflow creation steps:
* workflow trigger (on chat message, on schedule, etc..)
* add AI agent
* select chat model (open ai, anthropic, groq, etc..)
* provide open ai api key details assuming we selected open ai chat model (gpt-4o-mini for example)
* add memory (simple memory) to store all the session/context data.
* add tools as required (Wikipedia, calculator, gmail, weather, etc..)

#### Note:
we need to ensure we limit the LLM calls to reduce the costs and serve from memory as required. Not all the times, agent does the work in efficient way but we need to provide proper instructions so it can run more efficiently.

Example for efficient token usage:
* Add system message "fetch from memory if data is available and call the tool only when data is not in memory". This way Agent will follow instructions to reduce the calls to LLMs and hence reducing token usage and costs associated with it.
