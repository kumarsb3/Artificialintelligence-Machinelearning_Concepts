# Data frame

data frame is 2D structure, representation of a table. Its a collection of series where series is 1D structure

Data frame can n=be instantiated from a dictionary of series as columns OR using an array of lists as rows


```python
#from dictionary
data = {'one' : pd.Series([1., 2., 3.], index=['a', 'b', 'c']),
     'two' : pd.Series([1., 2., 3., 4.], index=['a', 'b', 'c', 'd'])}
df = pd.DataFrame(data)
df
```

```python

#JOINS

import pandas as pd
left = pd.DataFrame({
   'id':[1,2,3,4,5],
   'Name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'],
   'subject_id':['sub1','sub2','sub4','sub6','sub5']})
right = pd.DataFrame(
   {'id':[1,2,3,4,5],
   'Name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'],
   'subject_id':['sub2','sub4','sub3','sub6','sub5']})
middle = pd.DataFrame({
    'id':[6,7,8,9,10],
    'Name': ['abc','def','ghi','jkl','mno'],
    'subject_id':['sub6','sub7','sub8','sub9','sub10']})

print(left)
print("**************")
print(right)
print(middle)
```

```python
# prompt: we would like to concatenate the left and the right dataframe above. pls suggest a pandas code

import pandas as pd

# Assuming 'left' and 'right' are your DataFrames
# You can replace these with your actual DataFrames
# join on id column and inner join
#by default it will do inner join

merged_df = pd.merge(left, right, on='id', how='inner')
merged_df
```

```python
pd.merge(left,right,on='id')
```

```python

# join on 2 columns
pd.merge(left,right,on=['id','subject_id'])
```

```python
pd.merge(left, right, on='subject_id', how='outer', suffixes=('_l', '_r'))
```

```python
pd.merge(left, right, on='subject_id', how='right', suffixes=('_l', '_r'))
```

```python
pd.merge(left, right, on='subject_id', how='inner', suffixes=('_l', '_r'))
```

```python
pd.merge(left, right, on='subject_id', how='outer', suffixes=('_l', '_r'))
```

```python
pd.concat([left,right],ignore_index=True).sort_values(by='id').reset_index(drop=True)
```
